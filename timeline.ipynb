```python
import pandas as pd
from helpers.timeline import get_timeline
```

```python
milestones = pd.read_csv("milestones.csv")
milestones['datetime'] = pd.to_datetime(milestones.date)

spans = pd.read_csv("spans.csv")
spans['start_datetime'] = pd.to_datetime(spans.start)
spans['end_datetime'] = pd.to_datetime(spans.end)
```

```python
get_timeline(
    milestones,spans,'2022-06-05 12:00','2022-06-14',
    milestone_options=[
        {'index':0,'line_width':30.},
        {'index':1,'line_width':50.}
    ],
    filename='timeline-1.png'
)
```

```python
get_timeline(
    milestones,spans,'2022-06-14 12:00','2022-06-22',
    milestone_options=[
        {'index':10,'line_width':40.}
    ],
    span_options=[
        {'index':2,
        'annotation_point':'end_datetime','x_offset':15},
        {'index':0,'annotation_point':'end_datetime','x_offset':15}
    ],
    interval=24,
    filename='timeline-2.png'
)
```
